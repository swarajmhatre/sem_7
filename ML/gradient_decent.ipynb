```python
# import matplotlib as pyplot
from matplotlib import pyplot
import pandas as pd 
import sympy as sym
import numpy as np 
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient_decent(alpha_step,start, max_iter):
    x_list = list()
    x = start
    x_list.append(x)
    for i in range(max_iter):
        gradient = derivative(x)
        x = x - (alpha_step*gradient)
        x_list.append(x)
    return x_list
```

```python
def gradient_decent1(expr, alpha, start,max_iter):
    x_list = list()
    x = sym.symbols('x')
    grad = sym.Derivative(expr, x).doit()
    x_val = start;
    x_list.append(x_val);
    for i in range (max_iter):
        gradient = grad.subs(x, x_val);
        x_val = x_val - (alpha*gradient)
        x_list.append(x_val);
    return x_list
```

```python
start = 2
max_iter =30
alpha_step = 0.1
```

```python
x = gradient_decent(alpha_step, start, max_iter)
```

```python
x_cordinate = np.linspace(-5,5,100)
pyplot.plot(x_cordinate, objective(x_cordinate))
X_arr = np.array(x);
pyplot.plot(X_arr, np.array(objective(X_arr)), '.-', color='red')
pyplot.show()
```

```python
x = sym.symbols('x')
expr = (x+3)**2

x = gradient_decent1(expr, alpha_step, start, max_iter)
```
